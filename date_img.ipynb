```python
import cv2
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
img=cv2.imread(r'D:\DL Projects\OCR\Invoice.jpg')
img
```

```python
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
dic=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
import re
n=len(dic['text'])
for i in range(n):
    if dic['conf'][i]>50:
        if re.search(r'(\d{1,2}/\d{1,2}/\d{4})',dic['text'][i]):
            x, y, w, h = dic['left'][i], dic['top'][i], dic['width'][i], dic['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
